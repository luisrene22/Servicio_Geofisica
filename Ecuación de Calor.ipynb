Inicialmente, la barra está a una temperatura uniforme de 100 ° C, y luego ambos extremos se ponen en contacto con agua con hielo a 0 ° C. El calor sale de los extremos no aislados solamente. Su problema es determinar cómo variará la temperatura a medida que avanzamos a lo largo de la barra en momentos posteriores.

## La ecuación de calor parabólica (teoría)

Un hecho básico de la naturaleza es que el calor fluye de  un reservorio caliente a uno frío, es decir, de regiones de alta temperatura a regiones de baja temperatura.  Estas palabras en expresión matemática indican que la tasa de flujo de calor **H** a través de un material es proporcional al gradiente de la temperatura $T$ a través del material:

$$H=-k\nabla T(x,t)\quad\quad\quad(1)$$

donde $K$ es la conductividad térmica del material. La cantidad total de calor $Q(t)$ en el material en cualquier momento es proporcional a la integral de la temperatura sobre el volumen del material:

$$Q(t)=\int { dx{ C\rho  }(x)T(x,t) }\quad\quad\quad(2) $$

donde $C$ es el calor específico del material y $\rho$ es su densidad. Debido a que la energía se conserva, la tasa de disminución en Q con el tiempo debe ser igual a la cantidad de calor que sale del material. Después de que se alcanza este balance de energía y se aplica el teorema de divergencia, se obtiene $\textit{la ecuación de calor}$

$$\frac { \partial T(x,t) }{ \partial t } =\frac { k }{ C\rho  } { \nabla  }^{ 2 }T(x,t)\quad\quad\quad(3)$$

$\textit{la ecuación de calor}$ es una PDE ($\textit{3}$) parabólica con el espacio y el tiempo como variables independientes. La especificación de este problema implica que no hay variación de temperatura en las direcciones perpendiculares a la barra ($y$ y $z$), por lo que solo tenemos una coordenada espacial en el laplaciano:

$$\frac { { \partial  }T(x,t) }{ \partial t } =\frac { K }{ C\rho  } \frac { { \partial  }^{ 2 }T(x,t) }{ \partial { x }^{ 2 } }\quad\quad\quad(4) $$

Como se indica, la temperatura inicial de la barra y las condiciones de contorno son:

$$T(x,t=0)=100°C,T(x=0,t)=T(x=L,t)=0°C\quad\quad\quad(5)$$

## Solución: Expansión analítica

De manera análoga a la ecuación de Laplace, la solución analítica comienza con el supuesto de que la solución se separa en el producto de funciones de espacio y tiempo:

$$T(x,t)=X(x)\Gamma(t)\quad\quad\quad(6)$$

Cuando se sustituye la relación $(6)$ en la ecuación de calor $(4)$ y la relación resultante se divide por $X(x)\Gamma(t)$, el resultado es dos EDO no acopladas:

$$\frac { { d }^{ 2 }X(x) }{ d{ x }^{ 2 } } +{ k }^{ 2 }X(x)=0\quad,\quad\frac { { d }\Gamma (t) }{ dt } +{ k }^{ 2 }\frac { C }{ C\rho  } \Gamma (t)=0\quad\quad\quad(7)$$

donde $k$ es una constante aun por determinar. La condición de frontera que nos dice que la temperaturas son iguales a cero cuando $x=0$ requiere una funcion seno para $X$:

$$X(x)=Asinkx\quad\quad\quad(8)$$

La condición de frontera que nos dice que las temperaturas son iguales a cero en $x = L$ requiere que la función seno desaparezca allí:

$$sinkx=0\quad\Rightarrow \quad k={ k }_{ n }=\frac { n\pi  }{ L } \quad,\quad n=1,2,....\quad\quad\quad(9)$$

Se propone que la función de tiempo sea una exponencial en descomposición con $k$ en el exponente:

$$\Gamma (t)={ e }^{ { { k }_{ n }^{ 2 }t }/{ C\rho  } }\quad \Rightarrow \quad T\left( x,y \right) ={ A }_{ n }sin{ k }_{ n }x{ e }^{ { { k }_{ n }^{ 2 }t }/{ C\rho  } }\quad\quad\quad(10)$$

donde $n$ puede ser cualquier entero positivo y ${A}_{n}$ es una constante albitraria.

Debido a que $(4)$ es una ecuación lineal, la solucion más general es una superposición de productos ${X}_{n}(x){ \Gamma }_{ n }(t)$ para todos los valores de n:

$$T\left( x,t \right) =\sum _{ n=1 }^{ \infty  }{ { A }_{ n }sin{ k }_{ n }x{ e }^{ { { k }_{ n }^{ 2 }t }/{ C\rho  } } }\quad\quad\quad(11) $$

Los coeficientes de ${A}_{n}$ vienen determinadas por la condición inicial de que en el momento t = 0, toda la barra tiene temperatura $T = 100K:$

$$T\left( x,t=0 \right) =100\quad \quad \Rightarrow \quad \sum _{ n=1 }^{ \infty  }{ { A }_{ n } } sin{ K }_{ n }x=100\quad\quad\quad(12)$$

Si proyectamos las funciones seno determinadas por ${A}_{n}=\cfrac { 4{ T }_{ 0 } }{ n\pi}$ para n impares, y así 

$$T\left( x,y \right) =\sum _{ n=1,3,... }^{ \infty  }{ \cfrac { 4{ T }_{ 0 } }{ n\pi  } sin{ k }_{ n }x{ e }^{ { { k }_{ n }^{ 2 }t }/{ C\rho  } } }\quad\quad\quad(13)$$


```python
import matplotlib . pylab as p
from mpl_toolkits . mplot3d import Axes3D#Impotamos las librerias y herramientas que utilizaremos

print ("Por favor sea paciente, tengo paquetes para importar y puntos para trazar.")#Mensaje para el usuario

delta = 0.9#Declaramos un valor flotante para el tamaño de paso

n=int(input('ingrese numero n:'))#Declaramos un valor entero que lo proporcionara el ususario
                                 #al correr el programa, para saber hasta que numero se debe sumar.
L=1.#Es el valor del largo de la varilla metalica
KAPPA = 210.;SPH = 900.; RHO = 2700. #Conductividad, calor especifico y densidad


x = p.arange(0., 50., delta ) #Las lineas de codigo que van de la 14 a la 16 son el arreglo para graficar
y = p.arange(0., 50., delta ) # con ayuda de la libreria 'matplotlib', de tal forma que nos presentara un
X , Y = p.meshgrid(x, y)      #arreglo en R**3 con [x,y] en un intervalo de (0,50).


resultado = 0
for i in range(1,2*n+1):
    resultado += ((4*100)/(i*p.pi))*p.sin((i*p.pi)/(L))*(X)*p.exp((-((i*p.pi)/((L))**(2))*(KAPPA*(Y))/(RHO*SPH)))
print('',resultado)

#En las lineas de codigo que van de la 19 a 22 se realiza un ciclo for para realizar la operación de la suma
#de la ecuación (13) de este mismo notebook, el ciclo a barca un rango desde el numero 1 a un numero impar 
#que esta en función del valor 'n' que ingrese el usuario(como se menciono en la linea 8).
#Se agregado la opción de imprimir el resultado con el objetivo dever que tipo de dato arrojaba el resultado
#y ver si habia alguna congruencia en lo que se le pedia a la maquina,(aunque se puede omitir esa parte).


Z = resultado #Declaramos a la variable Z como el resultado obtenido en el ciclo 'for'


fig = p.figure ()
ax = Axes3D(fig)
ax.plot_surface (X,Y,Z)
ax.plot_wireframe(X, Y, Z, color = 'r') #La lineas de codigo de la 34 a la 41 son simplemente
ax.set_xlabel ( 'Posición' )            #son los comandos usuales para graficar con Matplotlib en R**3
ax.set_ylabel('Tiempo')
ax.set_zlabel('Temperatura')
p.show()
```

## Solución: Numérica por Diferencias finitas

Se determina el espacio y tiempo con una malla y se resuelven las soluciones en los sitios de la malla. Los sitios a lo largo de la parte superior con centros blancos corresponden a los valores conocidos de la temperatura durante el tiempo inicial, mientras que los sitios con centros blancos a lo largo de los lados corresponden a la temperatura fija a lo largo de los límites. Sin embargo, solo con las filas superior y lateral conocidas, terminaremos con un algoritmo que avanza en el tiempo una fila a la vez.

Como suele ser el caso con las PDE, el algoritmo se personaliza para la ecuación que se resuelve y para las restricciones impuestas por el conjunto particular de condiciones iniciales y de contorno. Con solo una fila de veces para empezar, usamos una aproximación de diferencia de avance para el tiempo derivado de la temperatura:

$$\frac { \partial T(x,t) }{ \partial t } \simeq \frac { T(x,t+\Delta t)-T(x,t) }{ \Delta t } $$ 

Debido a que conocemos la variación espacial de la temperatura a lo largo de toda la fila superior y los lados izquierdo y derecho, estamos menos limitados con la derivada del espacio que con la derivada del tiempo. En consecuencia, usamos la aproximación de diferencia central más precisa para la derivada de espacio:

$$\frac { { \partial  }^{ 2 }T(x,t) }{ \partial { x }^{ 2 } } \simeq \frac { T(x+\Delta x,t)+T(x+\Delta x,t)-2T(x,t) }{ { (\Delta x) }^{ 2 } }$$

La sustitución de estas aproximaciones en $(4)$ produce la ecuación de diferencia de calor

$$\frac { T(x,t+\Delta t)-T(x,t) }{ \Delta t } =\frac { K }{ C\rho  } \frac { T(x+\Delta x,t)+T(x-\Delta x,t)-2T(x,t) }{ \Delta { x }^{ 2 } }\quad\quad\quad(i) $$

Reordenamos $(i)$ en una forma en la que $T$ puede avanzar en $t$:

$${ T }_{ i,j+1 }={ T }_{ i,j }+\eta \left[ { T }_{ i+1,j }+{ T }_{ i-1,j }-2{ T }_{ i,j } \right] ,\quad \eta =\frac { K\Delta t }{ C\rho \Delta { x }^{ 2 } }\quad\quad\quad(ii) $$

donde $x=i\Delta x$ y $t=j\Delta t$.Este algoritmo es explícito porque proporciona una solución en términos de valores conocidos de la temperatura. Si intentáramos resolver la temperatura en todos los sitios de la malla simultáneamente, tendríamos un algoritmo implícito que nos obliga a resolver ecuaciones que involucran valores desconocidos de la temperatura. Vemos que la temperatura en el punto espacio-tiempo $(i,j+1)$ se calcula a partir de los tres valores de temperatura en un tiempo anterior j y en valores de espacio adyacentes $i\pm 1,i$.Comenzamos la solución en la fila superior, avanzamos en el tiempo que deseamos y mantenemos la temperatura en los extremos fijados en $0K$

```python
#EqHeat.py: solucion de 'la ecuacion de calor' por diferencias finitas, gráfica en R3
from numpy import *
import matplotlib.pylab as p #De la linea 2 de codigo a la 4 se importan las librerias y herramientas necesarias.
from mpl_toolkits.mplot3d import Axes3D

Nx = 101; Nt = 3000; Dx = 0.03; Dt = 0.9  #Se declara las dimensiones de la malla espacio terporal, al igual que 
                                          #el tamaño de sus pasos espacio-temporales.  
KAPPA = 210.; SPH = 900.; RHO = 2700.     #Conductividad, calor especifico y densidad
T = zeros((Nx,2),float); Tpl = zeros((Nx,31),float)

print('Trabajando, espera figura después de contar hasta 10')

for ix in range (1,Nx-1): T[ix,0] = 100.0;  # T inicial(temperatura inicial)
T[0,0] = 0.0 ; T[0,1] = 0.                  # primera y ultima T = 0(temperatura)
T[Nx-1,0] = 0. ; T[Nx-1,1] = 0.0
cons = KAPPA/(SPH*RHO)*Dt/(Dx*Dx);     #Constante
m = 1                                  #Contador

for t in range (1,Nt):
    for ix in range (1,Nx-1):
        T[ix,1] = T[ix,0] + cons*(T[ix+1,0]+T[ix-1,0]-2.*T[ix,0])
    if t%300 == 0 or t == 1:           # cada 300 pasos, se puede 'modificar'.
        for ix in range (1,Nx-1,2): Tpl[ix,m] = T[ix,1]
        print(m)
        m = m+1
    for ix in range (1,Nx-1): T[ix,0] = T[ix,1]
x = list(range(1,Nx-1,2))             #Grafica
y = list(range(1,30))
X , Y = p.meshgrid(x,y)

#De las linea de codigo de la 19 a la 29 se lleva acabo el metdo de diferencias finitas 
#En el caso particular de este codigo se realiza con cuatro ciclos 'for' uno principal y tres anidados
#y una operdor condicional 'if', todo en conjunto opera a partir de la euación discretizada (ii),
#que se encuentra en la sección 'Solución: Numérica por Diferencias finitas' en este notebook.

def functz(Tpl):
    z = Tpl[X,Y]
    return z

Z = functz(Tpl) 
fig = p.figure()
ax = Axes3D(fig)     #La lineas de codigo de la 36 a la 48 son simplemente
ax.plot_wireframe(X, Y, Z, color = 'b')#son los comandos usuales para graficar con Matplotlib en R**3
ax.set_xlabel('posición')
ax.set_ylabel('tiempo')
ax.set_zlabel('Temperatura')
p.show()
print('Terminado')
```

```python
# HeatCNTridia.py: solution of heate qtn via CN method

import matplotlib.pylab as p;#De la linea 3 de codigo a la 6 se importan las librerias y herramientas necesarias.
from mpl_toolkits.mplot3d import Axes3D;
from numpy import *;
import numpy;

Max = 51; n = 50; m = 50
Ta = zeros((Max),float); Tb = zeros((Max),float); Tc = zeros((Max),float)
Td = zeros((Max),float); a = zeros((Max),float); b = zeros((Max),float)
c = zeros((Max),float); d = zeros((Max),float); x = zeros((Max),float);
t = zeros((Max,Max),float)

def Tridiag(a,d,c,b,Ta,Td,Tc,Tb,x,n):
    Max = 51
    h = zeros((Max),float)
    p = zeros((Max),float)
    for i in range (1,n+1):
        a[i] = Ta[i]
        b[i] = Tb[i]
        c[i] = Tc[i]
        d[i] = Td[i]
    h[1] = c[1]/d[1]
    p[1] = b[1]/d[1]
    for i in range (2,n+1):
        h[i] = c[i] / (d[i]-a[i]*h[i-1])
        p[i] = (b[i]-a[i]*p[i-1]) / (d[i]-a[i]*h[i-1])
    x[n] = p[n]
    for i in range (n-1,1,-1): x[i]=p[i]-h[i]*x[i+1]

width = 1.0; height = 0.1; ct = 1.0
for i in range (0,n): t[i,0] = 0.0
for i in range (1,m): t[0][i] = 0.0
h = width / (n-1)
k = height / (m-1)
r = ct * ct * k / (h * h)

for j in range(1,m+1):
    t[1,j] = 0.0
    t[n,j] = 0.0
for i in range(2,n): t[i][1] = sin(pi*h*i)
for i in range(1,n+1): Td[i] = 2. + 2./r
Td[1] = 1.; Td[n] = 1.
for i in range(1,n): Ta[i] = -1.0; Tc[i] = -1.0;
Ta[n-1] = 0.0; Tc[1] = 0.0; Tb[1] = 0.0; Tb[n] = 0.0

print('trabajando')

for j in range(2,m+1):
    print(j)
    for i in range(2,n): Tb[i] = t[i-1][j-1] + t[i+1][j-1]+(2/r-2)*t[i][j-1]
    Tridiag(a,d,c,b,Ta,Td,Tc,Tb,x,n)
    for i in range(1,n+1): t[i][j] = x[i]
print('terminado')
x = list(range(1,m+1))
y = list(range(1,m+1))
X,Y = p.meshgrid(x,y)

def functz(t):
    z = t[X,Y]
    return z

Z = functz(t)

fig = p.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X,Y,Z,color='r')
ax.set_xlabel('t') 
ax.set_ylabel('x') 
ax.set_zlabel('T')
p.show()
        
```
